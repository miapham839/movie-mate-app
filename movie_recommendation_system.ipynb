```python
import numpy as np
import pandas as pd 
```

```python
#Read data from csv file
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv') 
```

```python
movies.head(2)
```

```python
movies.shape
```

```python
credits.head()
```

```python
#Merge the two datasets on the 'title' column
movies = movies.merge(credits,on='title')
```

```python
movies.head()
# budget
# homepage
# id
# original_language
# original_title
# popularity
# production_comapny
# production_countries
# release-date(not sure)
```

```python
#Remove irrelevant columns
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.head()
```

```python
type(movies['genres'][0])
```

```python
#Drop missing values
movies.dropna(inplace=True)
```

```python
import ast
```

```python
#Function to extract only the values of the 'name' key
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L 
```

```python
#Apply convert function to genres column
movies['genres'] = movies['genres'].apply(convert)
movies.head()
```

```python
#Apply convert function to keyword column
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

```python
movies['cast'][0]
```

```python
#modification of convert function (only get the first 3 names)
def convert_first_3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
            counter+=1
    return L 
```

```python
convert_first_3(movies['cast'][0])
```

```python
#apply convert_first_3 to 'cast' column to get the first 3 listed cast
movies['cast'] = movies['cast'].apply(convert_first_3)
movies.head()
```

```python
#fuction to get the director name
def get_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
```

```python
#apply get_director to the crew column. 
movies['crew'] = movies['crew'].apply(get_director)
```

```python
movies.sample(5)
```

```python
#Fuction to remove whitespace
def remove_space(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
#Apply function to remove space from cast, crew, genres, and keyword columns
movies['cast'] = movies['cast'].apply(remove_space)
movies['crew'] = movies['crew'].apply(remove_space)
movies['genres'] = movies['genres'].apply(remove_space)
movies['keywords'] = movies['keywords'].apply(remove_space)
```

```python
movies.head()
```

```python
#Tokenize the overview column
movies['overview'] = movies['overview'].apply(lambda x:x.split())
movies['overview'][0]
```

```python
#Create a new column called tags, which is overview, genres, keywords, cast, and crew combined
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
#create a new table with only movie_id, title, and tag column
new = movies.drop(columns=['overview','genres','keywords','cast','crew'])
new.head()
```

```python
new['tags'][0]
```

```python
new['tags'] = new['tags'].apply(lambda x: " ".join(x))
new.head()
```

```python
type(new['tags'][0])
```

```python
#Import scikit-learn library
import sklearn
```

```python
#Import TfidfVectorizer and create a TfidfVectorizer Object
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(max_features=5000,stop_words='english')
```

```python
#Create tfidf matrix for words in documents
tfidf_matrix = tfidf.fit_transform(new['tags'])
```

```python
print(tfidf_matrix)
```

```python
#Import NearestNeighors and instantiate a NearestNeighbors object 
from sklearn.neighbors import NearestNeighbors
model_knn = NearestNeighbors(metric='cosine', algorithm='brute')

#Fit the nearest neighbors estimator from the matrix
model_knn.fit(tfidf_matrix)
```

```python
#Recommendation function
def recommend(movie_title, n=5):
    # Get the index of the movie
    index = movies[movies['title'] == movie_title].index[0]

    # Find k nearest neighbors of that movie
    distances, indices = model_knn.kneighbors(tfidf_matrix[index], n_neighbors=n+1)

    for i in range(1, len(indices[0])):  # Skip the first (it's the same movie)
        print(f"{movies.iloc[indices[0][i]].title}")
```

```python
recommend('The Dark Knight', 5)
```
